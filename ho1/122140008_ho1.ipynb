<h3>Membuat Gelombang Sinus dan Cosinus</h3>

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat array t yang berkisar antara 0 hingga 2 dengan langkah 0.0001
t = np.arange(0, 2, 0.0001)

# Definisikan Sinyal
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi/2)
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)
```

<h3>Perbandingan Subplot</h3>

```python
# Membuat subplot 2x2 dengan nilai amplitudo = 2
fig, axs = plt.subplots(2, 2 , figsize=(8, 6))

# Plot setiap sinyal pada subplot
axs[0, 0].plot(t, y1)
axs[0, 0].set_title('y1 = 2 * sin(2 * π * 3 * t + 0)')
axs[0, 0].set_xlabel('t')
axs[0, 0].set_ylabel('y')
axs[0, 0].set_ylim(-2, 2)
axs[0, 0].grid(True)

axs[0, 1].plot(t, y2)
axs[0, 1].set_title('y2 = 1 * cos(2 * π * 4 * t + π/4)')
axs[0, 1].set_xlabel('t')
axs[0, 1].set_ylabel('y')
axs[0, 1].set_ylim(-2, 2)
axs[0, 1].grid(True)

axs[1, 0].plot(t, y3)
axs[1, 0].set_title('y3 = -1 * sin(2 * π * 5 * t + π/2)')
axs[1, 0].set_xlabel('t')
axs[1, 0].set_ylabel('y')
axs[1, 0].set_ylim(-2, 2)
axs[1, 0].grid(True)

axs[1, 1].plot(t, y4)
axs[1, 1].set_title('y4 = 0.5 * cos(2 * π * 6 * t + π)')
axs[1, 1].set_xlabel('t')
axs[1, 1].set_ylabel('y')
axs[1, 1].set_ylim(-2, 2)
axs[1, 1].grid(True)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

<h3>Pertanyaan Analisis</h3>

<p>1. Berapa Amplitudo dan Frekuensi masing-masing sinyal?</p>
<ul>
    <li>Sinyal 1 (y1): Amplitudo = 2 ; Frekuensi = 3</li>
    <li>Sinyal 2 (y2): Amplitudo = 1 ; Frekuensi = 4</li>
    <li>Sinyal 3 (y3): Amplitudo = 1 ; Frekuensi = 5</li>
    <li>Sinyal 4 (y4): Amplitudo = 0.5 ; Frekuensi = 6</li>
</ul>

<p>2. Bagaimana pergeseran fase mempengaruhi posisi gelombang? (Anda dapat mengubah nilai fase pada sinyal-sinyal yang telah dibuat sesuka anda)</p>
<ul>
    <li>Pergeseran fase +𝜙 menggeser gelombang ke kiri (lebih awal dalam waktu).</li>
    <li>Pergeseran fase -𝜙 menggeser gelombang ke kanan (lebih lambat dalam waktu).</li>
    <li>Pergeseran fase π membalikkan bentuk gelombang.</li>
<br>Jadi, pergeseran fase memengaruhi waktu mulai dan siklus gelombang tanpa mengubah frekuensi dan amplitudonya.
</ul>

<p>3. Bandingkan sinyal-sinyal dengan amplitudo yang berbeda dan diskusikan bagaimana amplitudo mempengaruhi tampilan gelombang.<br></p>
<p>Pada perbandingan subplot tiap sinyal yang tertera sebelumnya, dapat dilihat bahwa puncak dan lembah dari gelombang terdapat perbedaan. Yang mana gelombang dengan amplitudo lebih besar maka akan memiliki puncak yanag lebih tinggi dan lembah yang lebih dalam. Sedangkan untuk sinyal yang memiliki nilai amplitudo negatif, artinya gelombang yang dihasilkan terbalik terhadap sumbu horizontal namun tetap dengan ketinggian mutlak yang sama.</p>

<p>4. Bandingkan sinyal-sinyal dengan pergeseran fase yang berbeda dan diskusikan bagaimana pergeseran fase mempengaruhi tampilan gelombang.<br></p>
<p>Pada perbandingan subplot tiap sinyal yang tertera sebelumnya, dapat dilihat bahwa titik awal gelombang tiap sinyal berbeda-beda. Hal ini disebabkan oleh adanya pergeseran fase yang berbeda-beda.</p>
<ul>
    <li>Pada sinyal y1 (2 * sin(2π * 3t + 0)). Tidak terlihat pergeseran fase, ϕ = 0. Sehingga gelombang mulai dari nol.</li>
    <li>Pada sinyal y2 (1 * cos(2π * 4t + π/4)). Terjadi pergeseran fase sebesar π/4. Sehingga gelombang bergeser ke kiri sebesar T/8.</li>
    <li>Pada sinyal y3 (-1 * sin(2π * 5t + π/2)). Terjadi pergeseran fase sebesar π/2. Sehingga gelombang bergeser ke kiri sebesar T/4.</li>
    <li>Pada sinyal y4 (0.5 * cos(2π * 6t + π)). Terjadi pergeseran fase sebesar π. Sehingga gelombang terbalik dan bergeser setengah periode (T/2).</li>

<h3>Tugas Lanjutan</h3>

```python
# Membuat Sinyal baru kombinasi dari y1 dan y2
y5 = y1 + y2

# Membuat plot sinyal baru
plt.plot(t, y5)
plt.title('y5 = y1 + y2')
plt.xlabel('t')
plt.show()
```

<p>Gelombang di atas merupakan hasil kombinasi dari 2 frekuensi yang membentuk pola interferensi. Amplitudo maksimal dari kedua gelombang saling tumpang tindih. </p>
